```python
for i in range(12):
    epl[i] = epl[i].drop_duplicates(subset='Name')
    epl[i] = epl[i][['Name', 'Current Club']]
```

```python
epl[10]
```

```python
transaction = []

for i in range(11):
    transaction.append(pd.merge(epl[i], epl[i+1], on=['Name']))
```

```python
transfers = []

for i in range(10):
    transfers.append(transaction[i][transaction[i]['Current Club_x'] != transaction[i]['Current Club_y']])
```

```python
transfer_list = pd.concat(transfers)
```

```python
transfer_list = transfer_list.rename(columns = {'Current Club_x':'transfer_from', 'Current Club_y':'transfer_to'})
```

```python
transfer_list = transfer_list.drop('Name', axis=1)
```

```python
transfer_list.head()
```

```python
transfer_list = pd.get_dummies(transfer_list, columns=["transfer_from", "transfer_to"])
```

```python
transfer_list.head
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_transfers = apriori(transfer_list, min_support=0.01, use_colnames=True)
```

```python
rules = association_rules(frequent_transfers, metric="confidence", min_threshold=0.10)
```
